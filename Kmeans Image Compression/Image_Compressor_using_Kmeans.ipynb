```python
# import Kmeans_Python_Implementation; #ensure that this file is in the same directory
import numpy as np;
import matplotlib.pyplot as plt; import matplotlib.image as img;
import Kmeans_Python_Implementation as Kmeans;
# %matplotlib inline
```

```python
originalImage = plt.imread('Original_Images\\Me2_large.png');
X_img = np.reshape(originalImage, (originalImage.shape[0] * originalImage.shape[1], 3))
plt.imshow(originalImage);
```

```python
def TryKmeansFor(X_img, KArray, iterations = 10): #for each value of K from the KArray, we shall perform this 
    ListOfVals = [];
    for K in KArray:
        #initial_Centroids = Kmeans.initializeCentroids(X_img,K); 
        # centroids, indices = Kmeans.runKmeansAlgorithm(X_img,initial_Centroids,iterations); 
        centroids, indices = Kmeans.MultipleRandomStarts(X_img,K,3, iterations); 
        ListOfVals.append((centroids, indices));
    return ListOfVals; 
```

```python
Ks = [32]; iterations = 10;
output = TryKmeansFor(X_img,Ks,iterations); 
counter = 1; places = [(0,0),(0,1),(1,0),(1,1)]
fig,ax = plt.subplots(2,2, figsize = (8,8));
plt.axis('off');
ax[0][0].axis('off'); ax[0][0].imshow(originalImage);  ax[0][0].set_title('Original Image');#displays the original image in the first square
for (centroids,indices) in output:
    X_new = centroids[indices,:];
    X_new = np.reshape(X_new,originalImage.shape);
    ax[places[counter][0]][places[counter][1]].imshow(X_new); ax[places[counter][0]][places[counter][1]].set_title('compressed image with only %d colors'%Ks[counter-1]); 
    ax[places[counter][0]][places[counter][1]].axis('off'); 
    counter+=1;
    if(counter == 4):
        break; 
print(len(output)); 

fig.savefig('OutputImage2.png'); 
```
