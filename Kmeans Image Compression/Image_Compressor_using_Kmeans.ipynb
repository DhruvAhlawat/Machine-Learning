```python
# import Kmeans_Python_Implementation; #ensure that this file is in the same directory
import numpy as np;
import matplotlib.pyplot as plt; import matplotlib.image as img;
import Kmeans_Python_Implementation as Kmeans;
# %matplotlib inline
```

```python
originalImage = plt.imread('Original_Images\mona_lisa_128x128.png');
X_img = np.reshape(originalImage, (originalImage.shape[0] * originalImage.shape[1], 3))
plt.imshow(originalImage);
for i in range(0,20):
    print(X_img[i]);
```

```python
K = 2; iterations = 10; 
initial_Centroids = Kmeans.initializeCentroids(X_img,K); 
centroids, indices = Kmeans.runKmeansAlgorithm(X_img,initial_Centroids,iterations); 
```

```python
X_new = centroids[indices,:];
X_new = np.reshape(X_new,originalImage.shape);
fig,ax = plt.subplots(1,2, figsize = (8,8));
plt.axis('off');
ax[0].axis('off');
ax[0].imshow(originalImage); ax[0].set_title('Original Image'); 
ax[1].imshow(X_new); ax[1].set_title('compressed image with only %d colors'%K); 
```
